```python
y_train= traindf.iloc[:, 11].values
print(y_train)
```

```python
# encoding categorical net gain(dependent variable)
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelencoder_y = LabelEncoder()
y_train = labelencoder_y.fit_transform(y_train)
z=y_train.reshape(-1,1)
y_train=z
```

```python
features = ['id', 'realtionship_status', 'industry', 'genre', 'targeted_sex',
       'average_runtime(minutes_per_week)', 'airtime', 'airlocation',
       'ratings', 'expensive', 'money_back_guarantee']
X_train = traindf[features]
X_train.head()
X_test = testdf[features]
X_test.head()
```

```python
##ENCODING CATEGORICAL VARIABLES OF INDEPENDENT VARIABLES::
from sklearn.preprocessing import LabelEncoder,OneHotEncoder

#encoding genre
genre_le= LabelEncoder()
genre_labels= genre_le.fit_transform(X_train['genre'])
X_train['Genre_Label']= genre_labels

#encoding targeted sex
targesex_le= LabelEncoder()
targesex_labels= targesex_le.fit_transform(X_train['targeted_sex'])
X_train['TargeSex_Label']= targesex_labels

#encoding airlocation
airloc_le= LabelEncoder()
airloc_labels= airloc_le.fit_transform(X_train['airlocation'])
X_train['Airlocation_Label']= airloc_labels

#encoding expensive
expense_le = LabelEncoder()
expense_labels =expense_le.fit_transform(X_train['expensive'])
X_train['Expense_Label'] = expense_labels

#encoding industry,airtime,relationship using dummies and one-hot encoding approach
ind_dummy_features = pd.get_dummies(X_train['industry'])
rel_dummy_features = pd.get_dummies(X_train['realtionship_status'])
airtime_onehot_features = pd.get_dummies(X_train['airtime'])
money_dummy_features =pd.get_dummies(X_train['money_back_guarantee'])

X_train =pd.concat([X_train[['TargeSex_Label','Airlocation_Label','Expense_Label']],
                    rel_dummy_features,ind_dummy_features,airtime_onehot_features,money_dummy_features], axis=1)
X_train.head()
```

```python
#encoding the test set

#encoding genre
genre_le= LabelEncoder()
genre_labels= genre_le.fit_transform(X_test['genre'])
X_test['Genre_Label']= genre_labels

#encoding targeted sex
targesex_le= LabelEncoder()
targesex_labels= targesex_le.fit_transform(X_test['targeted_sex'])
X_test['TargeSex_Label']= targesex_labels

#encoding airlocation
airloc_le= LabelEncoder()
airloc_labels= airloc_le.fit_transform(X_test['airlocation'])
X_test['Airlocation_Label']= airloc_labels

#encoding expensive
expense_le = LabelEncoder()
expense_labels =expense_le.fit_transform(X_test['expensive'])
X_test['Expense_Label'] = expense_labels

#encoding industry,airtime,relationship using dummies and one-hot encoding approach
ind_dummy_features = pd.get_dummies(X_test['industry'])
rel_dummy_features = pd.get_dummies(X_test['realtionship_status'])
airtime_onehot_features = pd.get_dummies(X_test['airtime'])
money_dummy_features =pd.get_dummies(X_test['money_back_guarantee'])

X_test =pd.concat([X_test[['TargeSex_Label','Airlocation_Label','Expense_Label']],
                    rel_dummy_features,ind_dummy_features,airtime_onehot_features,money_dummy_features], axis=1)
X_test.head()
```
